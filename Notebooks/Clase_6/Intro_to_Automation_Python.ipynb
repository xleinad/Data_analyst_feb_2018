# Introducción a Automatización de Interfaz con Python

Primero instalamos y cargamos las librerías necesarias

```python
import pyautogui
```

## Controlando el mouse desde Python

Podemos pensar en la pantalla como un par de coordenadas (x,y) donde la esquina superior izquierda representa el (0,0) y los ejes x y y crecen hacia la derecha y hacia abajo, respectivamente

Podemos comenzar conociendo el tamaño de nuestra pantalla

```python
pyautogui.size()
```

```python
width, height = pyautogui.size()
```

Podemos obtener la posición actual del mouse con 

```python
pyautogui.position()
```

Probemos otro par de posiciones

```python
pyautogui.position()
```

```python
pyautogui.position()
```

Ahora veamos como mover el mouse

```python
pyautogui.moveTo(5,5)
```

Si queremos moverlo de manera más suave le podemos pasar el argumento duration

```python
pyautogui.moveTo(5,5,duration=1.5)
```

También podríamos mover el mouse en términos relativos

```python
pyautogui.moveRel(100,0)
```

```python
pyautogui.moveRel(0,-200)
```

Supongamos que quiero hacer click en el botón de guardar

```python
width, height = pyautogui.position()
```

```python
pyautogui.click(width,height)
```

```python
pyautogui.rightClick()
```

## Controlando el teclado desde Python

```python
pyautogui.click(100,100); pyautogui.typewrite("Holaaaaa")
```

Podemos mejorar lo anterior escribiendo

```python
pyautogui.click(100,100); pyautogui.typewrite(" Hola humano. Yo conozco todos tus secretos. Si... TODOS!",interval=0.15)
```

```python
pyautogui.KEYBOARD_KEYS
```

```python
pyautogui.press('esc')
```

Para apretar una combinación de teclas utilizamos hotkey()

```python
pyautogui.hotkey('esc','b')
```

```python
pyautogui.hotkey('cmd','f')
```

```python
# Probar en terminal
pyautogui.hotkey('shift','t')
```

## Screenshots y Reconocimiento de Imagen

```python
pyautogui.screenshot()
```

Para guardarlo, le pasamos como argumento el nombre del archivo

```python
pyautogui.screenshot('sh1.png')
```

```python
pyautogui.locateOnScreen('excel.png')
```

Vemos que nos regresa 4 valores en total, dos valores con las coordenadas (x,y) y otros dos valores con el tamaño de la imagen en cuestión

También podemos escribir

```python
pyautogui.locateCenterOnScreen('excel.png')
```

la cual nos da las coordenadas para el centro de la imagen reconocida
