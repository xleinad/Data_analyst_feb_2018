# **Taller Web Scraping**



## ¿Qué es Web Scraping?
#### -Es una técnica que sirve para extraer información de páginas web de forma automatizada.
#### -Nace a raíz de que la información contenida en estos sitios sólo puede ser vista mediante un explorador web, el cual no ofrece la funcionalidad de guardar estos datos para uso personal y que hace necesario copiar y pegar de forma manual los datos
#### -Hoy en día ya existen programas que automáticamente cargan y extraen estos datos usando parámetros definidos por el usuario; sin embargo no son fáciles de configurar y usar en la mayoría de los casos aparte de que involucran un alto costo.

Ejemplo: [WebHarvy](https://www.webharvy.com/articles/what-is-web-scraping.html).

## Librerias

```python
import pandas as pd
import numpy as np
import urllib.request as rq
import lxml.etree as et
from bs4 import BeautifulSoup
import requests
```

```python
url="http://www.artemis.bm/deal_directory/"
r=requests.get(url)
data=r.text
soup=BeautifulSoup(data,"lxml")
```

```python
print(soup.prettify())
```

```python
table=soup.find_all('table')[0]
rows=soup.find_all('tr')[2:]
```

```python
data={
    'Issuer': [],
    'Cedent': [],
    'Risks/Perils covered': [],
    'Size':[],
    'Date':[]
}

lr=len(rows)
lr
```

```python
for i in range(0,lr-1):
    data['Issuer'].append(rows[i].find_all('td')[1].get_text())
    data['Cedent'].append(rows[i].find_all('td')[2].get_text())
    data['Risks/Perils covered'].append(rows[i].find_all('td')[3].get_text())
    data['Size'].append(rows[i].find_all('td')[4].get_text())
    data['Date'].append(rows[i].find_all('td')[5].get_text())
        
ArtemisData=pd.DataFrame(data)
ArtemisData.head(10)
```

```python
complete_link=[]
for i in range(0,len(hyperlink_info)):
    link="http://www.artemis.bm"+hyperlink_info[i]
    complete_link.append(link)
    
complete_link[0]
```

```python
len(complete_link)
```

```python
ArtemisData['Hyperlink']=complete_link[0:lr-1]
ArtemisData.head(5)
```

```python
#Nueva forma de la base de datos
ArtemisData.shape
```

```python
#para cambiar nombre de columnas
ArtemisData=ArtemisData.rename(columns={'Issuer':'Issuer/SPV'})
ArtemisData=ArtemisData.rename(columns={'Cedent':'Cedent/Sponsor'})
ArtemisData.head(5)
```

```python
#para cambiar orden en el que aparecen
ArtemisData=ArtemisData[['Issuer/SPV', 'Cedent/Sponsor','Risks/Perils covered','Size','Date','Hyperlink']]
ArtemisData.head(5)
```

```python
placements=[]
for i in range(0,lr-1):
    url_placements=complete_link[i]
    r_placements=requests.get(url_placements)
    data_placements=r_placements.text
    soup_placements=BeautifulSoup(data_placements,"lxml")  
    deal_placements=soup_placements.find_all(['li','strong'])
    ldp=len(deal_placements)
    
    for s in range (1,ldp):
        if deal_placements[s].get_text()=='Placement / structuring agent/s:':
            a=deal_placements[s-1]
            b=a.get_text()
            placement=b[len("Placement / structuring agent/s: "):]
            placements.append(placement)
```

```python
deal_placements[0]
```

```python
risk_modellings=[]
for i in range(0,lr-1):
    url_risk=complete_link[i]
    r_risk=requests.get(url_risk)
    data_risk=r_risk.text
    soup_risk=BeautifulSoup(data_risk,"lxml")    
    deal_risk=soup_risk.find_all(['li','strong'])
    ldr=len(deal_risk)
    
    for s in range (1,ldr):
        if deal_risk[s].get_text()=='Risk modelling / calculation agents etc:':
            a=deal_risk[s-1]
            b=a.get_text()
            risk=b[len("Risk modelling / calculation agents etc: "):]
            risk_modellings.append(risk)  
```

```python
triggers=[]
for i in range(0,lr-1):
    url_trigger=complete_link[i]
    r_trigger=requests.get(url_trigger)
    data_trigger=r_trigger.text
    soup_trigger=BeautifulSoup(data_trigger,"lxml")    
    deal_trigger=soup_trigger.find_all(['li','strong'])
    ldt=len(deal_trigger)
    
    for s in range (1,ldt):
        if deal_trigger[s].get_text()=='Trigger type:':
            a=deal_trigger[s-1]
            b=a.get_text()
            trigger=b[len("Trigger type: "):]
            triggers.append(trigger)  
```

```python
ArtemisData['Placement / structuring agents']=placements
ArtemisData['Risk Modelling/Calculation Agents']=risk_modellings
ArtemisData['Trigger Type']=triggers
ArtemisData.head(5)
```
